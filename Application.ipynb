```python
class Vehicle:
    def __init__(self, brand, model):
        self.__brand = brand
        self.__model = model
    def move(self):
        return "Vehicle is Moving"
class Boat(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand, model)
    def move(self):
        return "Boat is Sailing"
class Plane(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand, model)
    def move(self):
        return "Plane is Flying"
v1 = Plane("Honda", "32")
v2 = Boat("Toyota", "42")
print(v1.move())
print(v2.move())
```

```python
import datetime as dt
import pickle
class Airport_management:
    def __init__(self, airportName, cityname, country):
        self.__airportName = airportName
        self.__cityname = cityname
        self.__country = country
        self.__flights = {}
        self.__passengers = {}
        self.__tickets = {}
        self.__employee = {}
        with open("Flight.csv", "rb") as file:
                while True:
                    try:
                        flight = pickle.load(file)
                        self.addFlights(flight)
                    except EOFError:
                        break
        with open("Passager.csv", "rb") as file:
                while True:
                    try:
                        passager = pickle.load(file)
                        self.addPassenger(passager)
                    except EOFError:
                        break
        with open("Ticket.csv", "rb") as file:
                while True:
                    try:
                        ticket = pickle.load(file)
                        self.addTicket(ticket)
                    except EOFError:
                        break
        with open("Employee.csv", "rb") as file:
                while True:
                    try:
                        employee = pickle.load(file)
                        self.addEmployee(employee)
                    except EOFError:
                        break
#displaying airport
    def displayAirportDetails(self):
        print (f"Airport Info\n------------\n{self.__airportName} {self.__cityname} {self.__country}\n")
#add flights
    def addFlights(self, flight):
        self.__flights[flight._Flight__FlightID] = flight
#display flights
    def displayFlights(self):
        rtn = "Flight Info\n------------\n"
        for all in self.__flights:
            rtn += (f"{self.__flights[all]}\n")
        print(rtn)
#add passenger
    def addPassenger(self, passenger):
        self.__passengers[passenger._Passenger__passengerID]=passenger
#display passenger
    def displayPassengers(self):
        rtn = "Passenger Info\n------------\n"
        for all in self.__passengers:
            rtn += (f"{self.__passengers[all]}\n")
        print(rtn)
#add tickets
    def addTicket(self, ticket):
        self.__tickets[ticket._Ticket__ticketID] = ticket
#displaying tickets
    def displayTickets(self):
        for all in self.__tickets:
            rtn = "Passenger, Flight and Ticket Info\n------------\n"
            rtn += (f"{self.__tickets[all]}\n")
            print(rtn)
#add employee
    def addEmployee(self, employee):
        self.__employee[employee._Employee__employeeID] = employee
#displaying employee
    def displayEmployees(self):
        rtn = "Employee Info\n------------\n"
        for all in self.__employee:
            rtn += (f"{self.__employee[all]}\n")
        print(rtn)
    def save(self):
            with open("Flight.csv", 'wb') as file:
                for c in self.__flights:
                    pickle.dump(self.__flights[c], file)
            with open("Passager.csv", 'wb') as file:
                for c in self.__passengers:
                    pickle.dump(self.__passengers[c], file)
            with open("Ticket.csv", 'wb') as file:
                for c in self.__tickets:
                    pickle.dump(self.__tickets[c], file)
            with open("Employee.csv", 'wb') as file:
                for c in self.__employee:
                    pickle.dump(self.__employee[c], file)
class Flight ():
    def __init__(self, flightID, airline, source, destination, date, startTime, reachTime):
        self.__FlightID = flightID
        self.__airline = airline
        self.__source = source
        self.__destination = destination
        self.__date = date
        self.__startTime = startTime
        self.__reachTime = reachTime
    def getFlightID(self):
        return(self.__FlightID)
    def __str__(self):
        return(f"{self.__FlightID} {self.__airline} {self.__source} {self.__destination} {self.__date} {self.__startTime} {self.__reachTime}")
    

class Passenger ():
    def __init__(self, passengerID, name, gender, ticketID = ""):
        self.__passengerID = passengerID
        self.__name = name
        self.__gender = gender
        self.__ticketID = ticketID
    def getPassengerID(self):
        return(self.__passengerID)
    def setTicket(self, str):
        Ticket(str.ticketID, str.passenger, str.flight, str.price)
    def __str__(self):
        return(f"{self.__passengerID} {self.__name} {self.__gender} {self.__ticketID}")
    

class Ticket():
    def __init__(self, ticketID, passenger, flight, price):
        self.__ticketID = ticketID
        self.__passenger = passenger
        self.__flight = flight
        self.__price = price
    def getTicketID(self):
        return(self.__ticketID)
    def __str__(self):
        return(f"{self.__passenger}\n{self.__flight}\n{self.__ticketID} {self.__passenger._Passenger__passengerID} {self.__flight._Flight__FlightID} {self.__price}\n")
class Employee():
    def __init__(self, employeeID, name, salary, designation):
        self.__employeeID = employeeID
        self.__name = name
        self.__salary = salary
        self.__designation = designation
    def getEmployeeID(self):
            return(self.__employeeID)
    def __str__(self) -> str:
        return(f"{self.__employeeID}  {self.__name}  {self.__salary}  {self.__designation}")
class AirportEmployee(Employee):
    def __init__(self, employeeID, name, salary, designation, dept):
        super().__init__(employeeID, name, salary, designation)
        self.__dept = dept
    def __str__(self) -> str:
        return (f"{Employee.__str__(self)}  {self.__dept}")
class AirlineEmployee(Employee):
    def __init__(self, employeeID, name, salary, designation, airline):
        super().__init__(employeeID, name, salary, designation)
        self.__airline = airline
    def __str__(self) -> str:
        return (f"{Employee.__str__(self)}  {self.__airline}")
        
    
changi = Airport_management("Changi", "Singapore", "Singapore")
changi.displayAirportDetails()
f1 = Flight('SQ123', 'Singapore Airlines', 'Singapore', 'Bangkok', dt.date(2023,6,1), dt.time(3,45,42), dt.time(6,30,45))
f2 = Flight('JAL832', 'Japan Airlines', 'Singapore', 'Tokyo', dt.date(2023,6,12), dt.time(1,30,00), dt.time(7,45,15))
changi.addFlights(f1)
changi.addFlights(f2)
changi.displayFlights()
p1 = Passenger('P001', 'John Smith', 'Male')
p2 = Passenger('P002', 'John Leong', 'Male')
changi.addPassenger(p1)
changi.addPassenger(p2)
changi.displayPassengers()
t1 = Ticket('T001', p1, f1, 450.55)
t2 = Ticket('T002', p2, f2, 470)
changi.addTicket(t1)
changi.addTicket(t2)
changi.displayTickets()
e1 = AirportEmployee("E001", "Patrick Wong", 4500.45, "Manager", "Ticketing")
e2 = AirlineEmployee("E002", "Jennifer Sim", 12300.33, "Pilot", "Eva Airlines")
changi.addEmployee(e1)
changi.addEmployee(e2)
changi.displayEmployees()
changi.save()
```
